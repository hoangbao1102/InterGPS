```python
import json
import matplotlib.pyplot as plt

# Đường dẫn đến file JSON chứa các giá trị loss
json_file = 'training_metrics.json'

# Đọc dữ liệu từ file JSON
with open(json_file, 'r') as f:
    data = json.load(f)

# Lấy các giá trị train_losses và val_losses từ dữ liệu
train_losses = data['train_losses']
val_losses = data['val_losses']

# Tạo list các epoch để làm trục x
epochs = list(range(1, len(train_losses) + 1))

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_losses, marker='o', linestyle='-', color='b', label='Train Loss')
plt.plot(epochs, val_losses, marker='o', linestyle='-', color='r', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Đường dẫn đến file JSON chứa các thông tin huấn luyện
json_file = 'training_metrics (3).json'

# Đọc dữ liệu từ file JSON
with open(json_file, 'r') as f:
    data = json.load(f)

# Khởi tạo các list để lưu thông tin từ JSON
iterations = []
running_losses = []
classification_losses = []
regression_losses = []

# Lặp qua từng entry trong dữ liệu JSON và lấy thông tin cần thiết
for entry in data:
    iterations.append(entry['iteration'])
    running_losses.append(entry['running_loss'])
    classification_losses.append(entry['classification_loss'])
    regression_losses.append(entry['regression_loss'])

# Điều chỉnh iterations để bắt đầu từ 0
iterations = list(range(len(classification_losses)))

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))

# Biểu đồ cho classification loss
plt.plot(iterations, classification_losses, label='Classification Loss')

# Biểu đồ cho regression loss
plt.plot(iterations, regression_losses, label='Regression Loss')

# Biểu đồ cho tổng running loss
plt.plot(iterations, running_losses, label='Running Loss')

plt.title('Training Losses')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
import json

# Đường dẫn đến file JSON chứa các thông tin huấn luyện
json_file = 'training_metrics (3).json'

# Đọc dữ liệu từ file JSON
with open(json_file, 'r') as f:
    data = json.load(f)

# Lấy thông tin về epoch và iteration của entry cuối cùng trong JSON
last_entry = data[-1]
current_epoch = last_entry['epoch']
current_iteration = last_entry['iteration']

# Tăng iteration lên 1
new_iteration = current_iteration + 1

# Tạo một entry mới với epoch và iteration như cũ, nhưng iteration mới là new_iteration
new_entry = {
    "epoch": current_epoch,
    "iteration": new_iteration,
    "classification_loss": 0.1,  # Thay đổi giá trị classification_loss theo nhu cầu
    "regression_loss": 0.2,      # Thay đổi giá trị regression_loss theo nhu cầu
    "running_loss": 0.3,         # Thay đổi giá trị running_loss theo nhu cầu
    "learning_rate": 1e-05       # Thay đổi giá trị learning_rate theo nhu cầu
}

# Thêm entry mới vào danh sách dữ liệu
data.append(new_entry)

# Lưu lại dữ liệu vào file JSON
with open(json_file, 'w') as f:
    json.dump(data, f, indent=2)

print(f"Added new entry with epoch {current_epoch} and iteration {new_iteration}.")
```
